# **Variables :-**

 * To Declare A Variable We Use **var** or **let** or **const**  keyword 

 * There Is Some Words Cannot Be Declared As A Variable Name For Example (var , if , function) 

 * Difference Between **var** , **let** and **const** :

      * **var :** 
      
           * variables can be re-declared and updated.

           * can lead to unintended behavior due to hoisting and global scope pollution

           * Can be declared without an initial value (defaults to undefined)

      * **let :** 
      
           * variables can be reassigned but not re-declared within the same block scope

           * prevent unintended variable hoisting and enables more predictable scoping behavior.

           * Can be declared without an initial value (defaults to undefined)


      * **const :** 
      
           * variables must be initialized with a value when declared and cannot be reassigned a new value later.

           * providing immutability and preventing accidental reassignment

# **Loops :-**


 * **For :**

   * **Syntax :**

     ```JS
     for (initialization; Condition; Iteration) {
       // Block Of Code
     }
     ```
             
     * **Initialization :** let i = 0

     * **Condition :** If True Continue The Loop   

     * **Iteration :** Increment or decrement i to break the loop

   * **Loop Control :**

     * **break :** 

       * immediately terminates the current loop, regardless of whether the loop's condition still true

       ```JS
        let products = ["Keyboard", "Mouse", "Pen", "Pad"]; 

        for (let i = 0; i < products.length; i++) {
          if (products[i] === "Pen") {
            break;
          }
          console.log(products[i])
        }
        /*
        "Keyboard"
        "Mouse"
        */
       ``` 

     * **continue :**

       * skips the remaining code within the current iteration of the loop and moves directly to the next iteration

       ```JS
        let products = [50, "Keyboard", "Mouse", 10, 20, "Pen", "Pad"];

        for (let i = 0; i < products.length; i++) {
          if (typeof products[i] === "number") {
            continue;
          }
          console.log(products[i]);
        }
        /*
        Keyboard
        Mouse
        Pen
        Pad
        */
       ``` 
        

     * **label :**

       * used in conjunction with break or continue to provide a target for them to jump to within a specific loop

       ```JS
        let products = ["Keyboard", "Mouse", "Pen", "Pad"];
        let colors = ["Red", "Black", "White"];

        mainLoop: for (let i = 0; i < products.length; i++) {
          console.log(products[i]);
          nestedLoop: for (let j = 0; j < colors.length; j++) {
            console.log(`-- ${colors[j]}`);
            if (colors[j] === "Black") {
              break mainLoop;
            }
          }
        }
        /*
        Keyboard
        -- Red
        -- Black
        */
       ```

# **Higher Order Function :**

 * functions that Take one or more functions as arguments  

 * can either use the provided function(s) internally or create and return a new function based on them

 * **map(callBackfFn, thisArg) :**

     * takes a callback function as an argument and applies it to each element of an array, returning a new array with the transformed elements

     * does not modify the original array; it creates a new one

     * **callBackFn(element, index, Array) :**

         * executed for each element in the array . It takes 3 arguments :

             1) element : The current element being processed 

             2) index : The index of the current element (Optional)

             3) array : The original array (Optional)


     * **thisArg :**

         * optional value to be used as **this** when executing the callback function  


     * **Example :**   

         ```JS
         let myNums = [1, 2, 3, 4, 5, 6];

         let add = myNums.map(function (element, index, array) {
         return element + element;
         });

         console.log(add); // [2, 4, 6, 8, 10, 12]
         ```
 



 * **filter(callBackFn, thisArg) :**   

     * used to create a new array containing only elements from the original array that pass a test defined by a callback function       

     * **Example :**

         ```JS
         let Nums = [10, 5, 12, 22, 13, 17];

         let filterEven = Nums.filter(function(ele) {
            return ele % 2 === 0;  
         })

         console.log(filterEven); // [10, 12, 22]
         ``` 

     * **Example :**

         ```JS
         // Filter & Map 

         let mix = "A13BS2ZX";

         let multipliedNums = mix
         .split("")
         .filter(function (el) {
             return !isNaN(parseInt(el));
         })
         .map(function (el) {
             return el * el;
         });

         console.log(multipliedNums); // [1, 9, 4] 
        
         ```   




 * **reduce(callBackFn, initialValue) :**

     * used to accumulate the elements of an array into a single value by applying a callback function to each element 

     * **callBackFn(Accumulator, currentValue, currentIndex, sourceArray) :**

         * **accumulator :** The accumulated value from the previous iteration, or the initialValue if provided    

         * **currentValue :** The current element being processed

         * **currentIndex :** The index of the current element (optional)

         * **array :** The original array (optional)


     * **initialValue :** 

         * optional value to be used as the initial value of the accumulator in the first iteration   


     * **Example :**

         ```JS 
          let nums = [10, 20, 30, 40]; 

          let adding = nums.reduce(function (acc, currV, currI, arr) {
            return acc + currV; 
          });

          console.log(adding); //100

          // InitialValue is 0 here , for example if InitVal = 10 the result of reducing will be equal to 110 not 100 

         ``` 

     * **Example :**

         ```JS

         // Filter & Reduce 
         
         let removeChars = ["@", "E", "@", "@", "L", "Z", "@", "@", "E", "R", "@", "O"];

         let word = removeChars
         .filter(function (el) {
             return el !== "@";
         })
         .reduce(function (acc, cur) {
             return acc + cur;  // Or Using Template Literals `${acc}${cur}`
         });

         console.log(word);  // ELZERO
         
         ``` 


 * **forEach(callBackFn(element, index, array), thisArg) :**

     * used to execute a provided function **once** for each element in an array   

     * it does not modify the original array and doesn't return anything **(returns undefined)**

# **Scopes :**

 * **Global :**

     * Variables declared outside of any function are considered globally scoped.

     * accessible throughout anywhere including inside functions


 * **Local :**

     * Also called Function Scope 

     * Variables declared inside a function are only accessible within that function and its nested functions if it exist 

     *  Not accessible from outside the function 


 * **Block :**

     * applies to code blocks enclosed within curly braces such as (if - for - switch)     

     * Variables declared using **let** or **const** within these blocks are only accessible within that specific block and any nested blocks within it 


 * **Lexical :**

     *  known as static scope

     * determines the accessibility of variables and functions based on their position in the code at the time of definition, not their execution location

# **Generator Function :**

 * Allows for pausing and resuming their execution during runtime ,It can yield multiple values one at a time

 * Returns special object (generator object)
 
 * **Syntax :**
  
     ```JavaScript
     function* functionName() {
        yield // to pause excution and return a value
     }
     ```

     * When the generator is resumed ( using **next()** method ), it picks up where it left off after the yield statement

     * Calling a generator function doesn't execute the function's code immediately. Instead, it returns a special iterator object

 * **Example :**

     ```JavaScript
     function* greetGenerator() {
     yield "Hello";
     yield "HI";
     alert("I am Here"); // Will not be excuted untill the next yield
     yield "Welcome";
     }

     let greeting = greetGenerator();

     console.log(typeof greeting); // Object

     console.log(greeting.next()); // {value: 'Hello', done: false}
     console.log(greeting.next()); // {value: 'HI', done: false}
     console.log(greeting.next()); // {value: 'Welcome', done: false} & the alert will be excuted
     console.log(greeting.next()); // {value: undefined, done: true}
     ```

 * **Delegate Generator Function :**

     ```JS
     function* generateN() {
         yield 1;
         yield 2;
     }

     function* generateL() {
         yield "A";
         yield "B";
    }

     function* generateAll() {
        yield* generateN();
        yield* generateL();
        yield* [3, 4];
     }

     let generator = generateAll();

     console.log(generator.next()); // {value: 1, done: false}
     console.log(generator.next()); // {value: 2, done: false}
     console.log(generator.next()); // {value: 'A', done: false}
     console.log(generator.next()); // {value: 'B', done: false}
     console.log(generator.next()); // {value: 3, done: false}
     console.log(generator.return()); // return() ends generating ==> {value: undefined, done: true}
     console.log(generator.next()); // {value: undefined, done: true}
     ```

# **JSON (JavaScript Object Notation) :**

 * text-based format for storing and transmitting structured data 

 * used for data exchange between web servers and web applications due to its simplicity and language independence

 * Does not Support Comments 


 * **Syntax :**

     * JSON syntax resembles JavaScript object literal syntax, using curly braces {} for objects and square brackets [] for arrays

     * Data is represented in key-value pairs, where keys are strings in double quotes and values can be strings, numbers, booleans, arrays, or null

     * Trailing commas ; are not allowed in JSON

     * Property names must be double-quoted strings

     * doesn't support data types like functions or symbols

     ```JSON
     {
        "string": "Sherif",
        "num": 50,
        "object": {
            "Egy": "Cairo",
            "KSA": "Ryade",
        },
        "array": ["JS", "PHP"],
        "boolean": false,
        "null": null 
     }
     ```

     * **JSON.parse()** convert a **JSON string** into a corresponding **JS object** literal   

     * **JSON.stringfy()** Converts a **JS value** (object, array, string, number) to a **JSON string**

 * **Asynchronous & Synchronous Programming :**

     * **Synchronous :**

         * Operations Runs in Sequence, Each Operation must wait for the previous one to complete 

         * It follows a blocking pattern.

         * **JS** is a Single-threaded language that executes instructions sequentially, one after the other 

             * **Example :**

                 ```JS 
                 console.log("Before");
                 window.alert("Operation"); 
                 console.log("After"); // This Line will not be be excuted untill you click OK on the alert Message  
                 ```

     * **Asynchronous :**

         * Operations run in parallel, multiple operations can occur at the same time

         * uses a non-blocking approach. It can initiate multiple tasks and continue execution without waiting for each one to finish entirely

             * **Example :**

                 ```JS
                 console.log("1");
                 setTimout(() => console.log("Operation"), 3000);
                 console.log("2");
                 ```

 * **Call Stack :**

     * Manages the execution flow of function calls in your JS code  

     * Call Stack operates on a principle called **LIFO (Last In, First Out)**

     * When a function is called, it's pushed onto the call stack, becoming the active function, then the code within the function executes , Once the function finishes execution (returns a value or reaches the end), it's popped off the stack

     * **Example :**

         ```JS
         function greet(name) {
             console.log(`Hi ${name}`);
         }

         function sayHi() {
             greet("Sherif");
             console.log("Are You OK ?");
         }

         sayHi();

         /*
         Hi Sherif
         Are You OK ?
         */

         /*
         ***********Explaination**********

         => When sayHi() is Called it's added to the Call Stack
         => Inside sayHi() , greet() is called with argument "Sherif"
         => The call stack now has greet() on top, followed by sayHi(). greet() becomes the active function
         => greet() executes and logs "Hi Sherif" to the console
         => greet has no more statements, it finishes execution and is popped off the Call Stack
         => Control returns to sayHi(), which is now the top function on the call stack
         => sayHi() continues execution and logs "Are You OK ?" to the console

        */
         ```

 * **CallBack Queue & Event Loop :**

     * Handles Asynchronous Operations without blocking the main thread , When an asynchronous operation is triggered in your JS code, the associated callback function is placed in the callback queue   

     * **Role Of The CallBack Queue :**

         * **1- Initiating the Operation :**
         
             * When JS code triggers an asynchronous operation
         
         * **2- Callback Registration :** 
         
             * The function associated with the operation **(the callback)** is registered and pushed onto the **callback queue**. This signifies the function's readiness to handle the response or result when it becomes available.
         
         * **3- Main Thread Continues :** 
         
             * The main thread of execution in JS doesn't wait for the asynchronous operation to finish. It moves on to other tasks in your code.
        
         * **4- Event Loop Takes Over :** 
         
             * Manages two things :

                 * **a- The call stack :**  Which manages currently executing functions

                 * **b- The callback queue :** which holds callbacks waiting for asynchronous operations to finish   
         
         * **5- Callback Execution :** 

             * When the call stack becomes empty (no more functions are actively executing), the event loop takes an action:

                 * It retrieves the first callback function (FIFO) from the callback queue.
                 
                 * It pushes this callback function onto the call stack.
                 
                 * The callback function is finally executed, allowing it to process the response or result from the completed asynchronous operation

 * **CallBack Hell :**

     * refers to a situation where you nest multiple callback functions within each other to handle asynchronous operations. This can lead to complex, difficult-to-read, and error-prone code


     * **Example :**

         ```JS

             console.log("Download Photo From URL");
             setTimeout(() => {
                 console.log("Resize Photo");
                 setTimeout(() => {

                     setTimeout(() => {
                         console.log("Show The Photo In Website");
                     }, 1000);
                 }, 1000);
             }, 1000);
         }, 1000);
         ```

     * **Callback Hell Solutions :**

         * **1- Promises :**

             * more structured and readable approach to handling asynchronous operations compared to traditional callback hell

             * provide a mechanism for dealing with the inherent uncertainty of when an asynchronous operation will complete. 

             * **Promise() Object :**

                 * represents the eventual completion (or failure) of an asynchronous operation

                 * A callback used to initialize the promise , Called **Executer Function** 
                 
                     * This callback is passed two arguments : 
                     
                         * **1- resolve :** used to resolve the promise with a value or the result of another promise
                         
                         * **2- reject :** used to reject the promise with a provided reason or error

                 * **Promise States :**

                     * **1- Pending :**

                         * The initial state, indicating the operation is ongoing  

                     * **2- Fullfilled :**

                         * The operation succeeded, and a result value is available

                     * **3- Rejected :**

                         * The operation failed, and an error reason is available 

                 * **Creating A Promise :**

                     ```JS
                     const myPromise = new Promise((resolve, reject) => {
                         // Executor function (performs asynchronous operation)
                         if (/* condition for success */) {
                             resolve("Success!"); // Fulfill the promise with a result
                         } else {
                             reject(new Error("Failed!")); // Reject the promise with an error
                         }
                         });
                     ```


                 * **Thenable :**

                     * An object that adheres to the Promise API, meaning it has a **then()** method for handling successful results 
                        
                         * **then()** method takes a callback function that executes when the promise is **fulfilled**. You can chain multiple **.then()** methods to handle results from multiple asynchronous operations.
                     
                     * **catch()** method for handling errors. Promises are inherently thenables  

                         * The **catch()** method takes a callback function that executes when the promise is **rejected**. It's used for centralized error handling  

                     * **finally()** method offers a way to execute code regardless of whether the promise is fulfilled (resolved) or rejected    

                     * **Handling Promise Resolution/Rejection :**

                         ```JS
                         myPromise
                             .then((result) => {
                                 console.log("Success:", result); // Process the successful result
                             })
                             .catch((error) => {
                                 console.error("Error:", error); // Handle the error
                             });
                             .finally(console.log("Anything");)
                         ``` 

             * **Example :**

                 ```JS
                 const myPromise = new Promise((resolveFn, rejectFn) => {
                     let applicants = ["Sherif", "Mohamed", "Mahmoud", "Ahmed"];
                     if (applicants.length === 4) {
                         resolveFn(applicants);
                     } else {
                         rejectFn(Error("Not Suitable For Job"));

                 myPromise
                     .then((resolveValue) => {
                         resolveValue.length = 2;
                         return resolveValue;
                     })
                     .then((resolveValue) => {
                         resolveValue.length = 1;
                         return resolveValue;
                     })
                     .then((resolveValue) => {
                     console.log(`The Choosen Applicant Is ${resolveValue}`);
                     })
                     .catch((rejectedReason) => console.log(rejectedReason))
                     .finally(console.log("The Interview Is Done")); 
                 ```

 * **Web APIs :**

     * act as intermediaries between your JS code and external data sources or functionalities 

     * **Types Of Web APIs :**

         * **Server-Side Web APIs :**

             * Reside on web servers and offer data or functionalities to web applications through well-defined endpoints (URLs)

             * Communication occurs using HTTP requests (GET, POST) and responses (with data in formats like JSON or XML)


         * **Client-Side Web APIs :**

             * Also Called **Browser APIs** , It **Acts** As a second Thread

             * Built into web browsers, allowing JavaScript (JS) code to access browser features and functionalities  

             * Provide programmatic control over browser behavior, like making network requests (Fetch API), accessing geolocation (Geolocation API), or manipulating DOM elements      



     * **How The Call Stack and The Web APIs Work Together ?**

         * **1- Initiating the Request :**  

             * JS code makes a call to a client-side Web API(Fetch API) to interact with a server-side Web API . This call is added to the call stack

         * **2- Fetching Data :**

             * Fetch API initiates a network request to the specified server-side API endpoint (URL)     

             *  Since the network request might take time (waiting for a response from the server), the JS code doesn't block. It continues execution or moves on to other tasks  ==> **(Asynchronous Programming)**

             * Fetch API add The **CallBack** to **CallBack Queue**

         * **3- Server Processing :**

             *  The server-side API processes the request, retrieves or manipulates data, and sends a response back    

         * **4- Handling The Response :**

             * Once the server responds, the network request completes, and control returns to the JS code in the call stack 

             * The Fetch API function parses the response (often JSON) and makes the data usable within your web application

 * **Fetch API :**

# **AJAX (Asynchronous JS And XML) :**

 * technique used in web development to make asynchronous requests from a web page to a server 

 * allows web pages to update their content dynamically without requiring a full page reload, providing a more seamless and responsive user experience 

 * **AJAX Components :**

     * **1- XMLHttpRequest() :**

         * used to make requests to the server , It's a browser-built object that allows communication between the client-side (web page) and the server

     * **2- JavaScript :**

         * interacts with the XHR object to send requests, receive responses, and manipulate the retrieved data to update the page content dynamically  


 * **AJAX WorkFlow :**

     * **1- Event Trigger :**

         * An event on the web page triggers an AJAX request (any event that initiates communication with the server)         


     * **2- Creating XHR Object :** 

         * JS Code Creats XHR Object 

             ```JS
             let myRequest = new XMLHttpRequest()
             ```

     * **3- Configuring The Request :**

         * By Opening The Connection Using **open()** method

             * **open(HTTP Method(String), URL(String), Async(Boolean)) :**

                 * **HTTP Method :**

                     * Specify the HTTP method (GET, POST, PUT, etc.) to indicate the type of operation to be performed on the server

                 * **URL :**

                     * URL of the server-side script that will handle the reques 

                 * **async :**

                     * Indicates Whether the Connection is Asynchronous or not (Usually true for AJAX) 

     * **4- Sending The Request :**

         * Using **send(data)** Method      

     * **5- Handling The Response :**

         * Set up the **onreadystatechange** Event Listener, This function is called multiple times as the request progresses through different states   

             * **Ready States :**

                 * **0 (Unsent) :** 
                      
                     * The request hasn't been configured using the **open()** method yet
                     
                 * **1 (Opened) :** 
                     
                     * Connection with the server has been established using **open()**
                     
                 * **2 (Headers Recieved) :**

                     * he server has sent response headers back to the client, but the response data is not yet received

                 * **3 (Loading) :** 

                     * The server is transferring the response data to the client
                     
                 * **4 (Done) :** 

                     * The request has completed. The server has sent all the data, and the response is available   

     * **6- Processing The Response :** 

         * When the request completes successfully **(status code 200)**, the **onreadystatechange** function checks the **readyState** property of the XHR object. If it's **4 (request finished)**, the response data can be accessed using the **responseText** or **responseXML** properties (depending on the expected data format, usually JSON or XML)  

             * **Status Code :**

                 * code conveys the overall status of the request and whether it was successful or not

                     * **200 (OK) :**

                         * The request was successful, and the server has sent the requested data in the response body

                     * **404 (NotFound) :**

                         * The server couldn't find the requested resource

                     * **403 (Forbidden ) :**

                     * **400 (BadRequest) :**

                         * The server couldn't understand the request due to invalid syntax or missing information              

         * JS code then parses **(JSON.parse())** and uses the received data to update the web page content dynamically      


     * **7- Error Handling :**
