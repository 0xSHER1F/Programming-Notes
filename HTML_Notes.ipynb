# **HTML (HyperText Markup Language) :** 

 * Used to form the foundation of web pages 

 * Main HTML Tags are :

      1) **head** : 
      
           * acts as a container for metadata about the webpage, This information is not directly displayed on the page but is crucial for browsers and search engine

                * **Main Elements Inside **head** tag :**

                     * **< title > :**

                          * defines the title of the webpage, displayed on the browser tab and search engine results

                     * **< meta > :**

                          * A Self-Closing Tag, provides various metadata attributes like : 
                          
                               * **Character Encoding :**

                                    * Defines the character encoding used in the webpage

                                      ```html
                                      <meta charset="UTF-8">
                                      ```

                               * **Page Description :**

                                    * Provides a brief description of the webpage for search engines

                                      ```html
                                      <meta name="description" content="WebSite About Programming" />
                                      ```

                               * **Viewport :** 

                                    * Controls how the webpage is displayed on various devices, particularly for responsive design

                                      ```html
                                       <meta name="viewport" content="width=device-width, initial-scale=1.0" />
                                      ```          
                               
                               * **Author :**

                                    * Specifies the author of the webpage

                                      ```html
                                       <meta name="author" content="S" />
                                      ```        

                     * **< style > :**

                          * Embeds (CSS) within the HTML document, controlling the visual appearance and presentation of your webpage elements  

                     * **< scritp > :**

                          * Allows to embed scripts, most commonly written in JS, to manipulate the content, style, and user experience of the webpage  

                     * **< link >** : 

                          * Self-Closing Tag , Connects the webpage to external resources that enhance its presentation or functionality 

                               ```html
                               <link rel="stylesheet" href="style.css"/> 
                               ```

                               * **rel :** 

                                    * Defines the relationship between the current document and the linked resource 

                                    * In Previous Example : set to "stylesheet", indicating that the linked resource is a CSS stylesheet     

                               * **href :**

                                    * specifies the location (URL) of the external resource being linked 

                                    * In Previous Example : specifies the path to the external stylesheet file, "style.css", which is located in the same directory as the HTML file                        

      2) **body** : encloses the main content that users see on the webpage 

           * **Main ELements Inside bode Tag :**

                * **Heading Tags :**

                     * defined by the tags from **< h1 >** to **< h6 >**, are fundamental elements used to structure your webpage's content and improve readability

                          * **h1 :** The main heading of your webpage, For Best Practice **h1** Should Be Used Only One To Time

                          * **h2 :** Subheadings within the main topic, used for major sections or breakdowns

                          * **h3 to h6 :**  Further subheadings for subsections, becoming progressively more specific as you go down the hierarchy

                     * They play a great role in Search Engine Optimization (SEO)

                * **Paragraph Tag < p > :**

                     * defines a paragraph of text, separating it visually from other elements on your webpage

                     * creates a **block-level** element that groups a section of related text content 

                     * Browsers add **margins** before and after each paragraph  

                     

 * **Basic Example :**

      ```html
      <!DOCTYPE html> <!-- (html) indicates that we use HTML vestion 5-->
      <html lang="en">
      <head>
           <title>My First Webpage</title>
           <meta charset="UTF-8" />
           <meta name="viewport" content="width=device-width, initial-scale=1.0" />
      </head>
      <body>
           <h1>Hello, World!</h1> 
           <p>This Is A Basic WebPage</p>
      </body>
      </html>
      ```

 * **!DOCTYPE :**

      * Tells the browser what version of HTML the document is written in, nowadays probably we use HTML5 doctype     

      * Without <!DOCTYPE>, the browser might use a less-predictable **"quirks mode"** for rendering, which can lead to inconsistencies

      * Can Be Written in LowerCase & UpperCase 

 * **HTML Entites :**

      * Way to represent characters that have special meaning in HTML itself, or that are difficult to type using a standard keyboard 

      * **For More Info =>** https://www.w3schools.com/html/html_entities.asp

 * **Formatting Elements :**  

      * **< b > & < strong > :** 

           * Make The Text Bold 

           * **< b >** applies bold formatting to the text without any inherent semantic meaning (Just A Styling)

           * **< strong >**  indicates strong importance within the document. It makes the text bold visually, but more important  

      * **< i > & < mark > :**

           * Presents text in italics, with **< em >** implying a stressed or technical term   

      * **< u > :**

           * Underlines text, though it's rarely used for semantic reasons

      * **< small > :**

           * Make text in a smaller font size compared to the surrounding text   

      * **< del > & < s > :**

           * Strikes a line through text, with **< del >** implying deleted content 

      * **< ins > :**

           * signifies inserted content within a document. It's useful for highlighting changes or additions made to text   

      * **< sup > and < sub > :**

           * Formats text to appear slightly above **(< sup >)** or below **(< sub >)** the baseline 

      * **< span > :**

           * an inline elemenpt, Group inline content for stylings purposes using CSS
s
                ```html
                <p>Iam <span style="color: red">Red</span> And He Is <span style="color: blue">Blue</span></p>
                <!--
                Red Word its Color Become red , Blue Word its Color Become blue
                -->
                ```

      * **< hr > :**

           * Self-Closing Tag , represents a thematic break , It displays pas a horizontal line across the webpage, visually separating content sections 

                ```html
                <p>Section one</p>
                <hr/>
                <p>Section Two</p>
                ```

      * **< br > :**

           * Self-Closing Tag , represents a line break , It forces the content to start on a new line within a block-level element 

      * **< div > :** 

           * **The Most Important Element**, It is a block-level element that acts as container for grouping content and applying styles

           * We can target these divs using classes, IDs, or other selectors in your CSS code , We Can Control (Background Color - Padding&Margins - Borders - Text Alignment)


      * **< q > :**

           * **inline Quotation** Used to represent a short inline quotation within a paragraph or sentence

      * **< blockquote > :**

           * **block-level quotation** Used to represent a longer quotation that is visually distinct from the surrounding text

      * **< wbr > :**

           * **Word Break** , Inserts a potential line break opportunity within a word 

           * Useful for preventing excessively long words from overflowing onto a single line and causing layout issues

      * **< bdi > :**

           * **Bidirectional Isolation** , Isolates a piece of text for bidirectional text rendering 

      * **< code > :**

           * Represents a fragment of computer code within a text      


      * **< pre > :**

           * Formats the enclosed text in a preformatted manner      


      * **< iframe > :**

           * Embeds another HTML document within the current document, creating a nested browsing contex

# **Links :**


 * **Links :**

   * Allow users to navigate between webpages and resources on the internet

   * created using the **< a >** tag , **a** stands for Anchor

   * **Example :** 

     ```htm
     <a href="https://example.com">Click Here</a>
     ```

     * Text **"Click Here"** When a user clicks on it, the browser will navigate to the webpage at the specified URL **"https://example.com"**

     * **href** Can Be An Email Address For Contacting 

       ```htm
       <a href="mailto:sherifsaady32@gmail.com">Click Here</a>
       ```

   * **Main Attributes of < a > :**

     * **target :** 

       * defines how the linked webpage should open in the browser window  

         ```htm
         <a href="https://example.com" target="_blank">Click Here</a>
         ```
           * Opens the linked content in a new browser tab or window

         ```htm
         <a href="https://example.com" target="_self">Click Here</a>
         ``` 
           * Opens the linked content in the current window (default)

         ```htm
         <a href="https://example.com" target="_top">Click Here</a>
         ```   

           * Opens the linked content in the full body of the window, replacing the current frame content

         ```htm
         <a href="https://example.com" target="_parent">Click Here</a>
         ``` 
           * Opens the linked content in the parent frame (if the current page is within a frame) 

     * **title :**

       * specifies advisory information about the link, appearing as a tooltip when users hover over the link with their mouse     

         ```htm
         <a href="https://FaceBook.com" target="_blank" title="Go To FaceBook">Click Here</a>
         ```

     * **download :**

       * specifies that the linked resource should be downloaded when the user clicks on the link. It's often used for downloadable files like PDFs or documents

# **Images :**

 * images are embedded using the **< img >** tag , **< img >** is a Self-closing tag

     ```html
     <img src="image URL or Its Path" alt="Alternative Text">
     ```

     * **src :**

         * specifies the path (URL) to the image file you want to display , It Can Be :

             * **Absolute URL :** A complete web address points to the image hosted on a remote server

                 ```html
                 <img src="https://hdqwalls.com/eclipse-abstract-5k-wallpaper" alt="Alternative Text">
                 ```

             * **Relative URL :** A path relative to the location of your HTML file. For instance, if the image is in the same folder as your HTML file

                 ```html
                 <img src="image.jpg" alt="Alternative Text">
                 ```


     * **alt :**

         * specifies alternative text that describes the image content

         *  This text is displayed in case the image cannot be loaded for some reason

# **Lists :**

 * allow you to present items in a structured and visually distinct manner

 1) **< ul > Unordered Lists :**

    * Used for groups of items where the order doesn't necessarily matter, displayed with bullet points   

    * **Creating A List Using < ul > :**

       ```html
       <ul>
         <li>Hi</li>
         <li>Hello</li>
         <li>Welcome</li>
       </ul>  

       <!--
            * Hi
            * Hello
            * Welcome
       -->
       ``` 
     
      * **< li > (list item)** element for each item inside **< ul >**

 2) **< ol > Ordered Lists :**

    * Used for sequences where the order of items is important, displayed with numbered items (Default) 

    * **< ol > attributes :**

       * **reverse :**
        
          * Reverses List Ordering

       * **start :**

          * Ordering Starts From Specific Number

          ```html
          <ol start="4">
            <li>First</li>  <!-- <li value="4"> Equivalent to <ol start="4">-->
            <li>Second</li>
            <li>Third</li>
          </ol>  
          <!--
               4. First
               5. Second
               6. Third
          -->
          ```  

       * **type :**

          * defines the type of numbering scheme used for the list items

             * **1 (Default):** displays a standard numerical list using decimal numbers 

             * **a :** uses lowercase alphabetical letters (a, b, c, d, ...) for numbering

             * **A :**  uses uppercase alphabetical letters (A, B, C, D, ...) for numbering

             * **I :**  uses uppercase Roman Numerals (I, II, II) for numbering     

 3) **< dl > Description Lists :**

    * create lists of terms and their associated descriptions, often used for glossaries and dictionaries

    * **Consists OF :**

       * **< dl > Container Tag :**  Marks the beginning of the description list

       * **< dt > Description Term :** Each term within the list is defined using it, it represents the item being defined

       * **< dd > Description Definition :** The corresponding description for each term is enclosed within it. it provides the explanation associated with the term 

    * **Example :**

       ```html
       <dl>
          <dt>HTML</dt>
          <dd>HyperText Markup Language. Foundation of web pages</dd>
       </dl>
       ```

 * **Nested Lists :**

    *  helpful for creating subcategories or hierarchical structures within the list

    * **Example :**

       ```html
       <ul>
         <li>Main Item
            <ul>
               <li>SubItem 1</li>
               <li>SubItem 2</li>
            <ul>
         </li>
       </ul>
       ```

# **Table :**
 
 * present data in a structured format of rows and columns 

 * **Main Elements Of Table :**

      * **< table > :** The foundation for a table, which defines the start of the table element

      * **< tr > :** Each row of data within the table is defined using this tag

      * **< td > :** content within each table cell are enclosed by the **< td >** tag. There can be multiple **< td >** tags within a **< tr >** to represent cells in that row

      * **< th > :** used for table headers. It displays the header text in bold

      * **< thead > :** Groups the header row(s) of the table

      * **< tbody > :** Groups the body content of the table 

      * **< tfoot > :** Group the footer row(s) of the table (often used for summaries)  

      * **< caption > :** Defines a caption (title) for the table, displayed above the table
      

 * **colspan & rowspan :** used with **< td >** to specify that a cell should span multiple columns/rows, respectively


 * **Example :**

# **Semantic Elements :**

# **Audio & Video:** 

 * **Audio :**

     * embedded into web pages using the **< audio >** tag , which provides a standardized way to include sounds and music

     * **Example :**

         ```html
         <audio>
            <source src="Audio.mp3" type="audio/mpeg">
            Browser Does Not Support This Audio <!--Displays When the Browser doesnot Support this audio type-->
         </audio>   
         ```

         * **< source > :** acts as an alternative media source provider for elements like < audio >
           
             * **src :** specifies the URL or path to the media file

             * **type :** defines the MIME type of the media file ("audio/mpeg" for MP3) 

     * **Main Attributes :**

         * **controls :**  adds a default playback interface with buttons for play, pause and volume

         * **autoplay :**  allows the audio to start playing automatically when the web page loads

         * **loop :** makes the audio play continuously, restarting after it finishes

         * **muted :** mutes the audio by default 


 * **Video :**    

     * **< video >** is used to embed video content into the web page .
     
     * It provides a standardized way to display videos without relying on external plugins 

     * **Example :**

         ```html
         <video>
            <source src="Video.mp4" type="video/mp4">
            Browser Does Not Support This Video <!--Displays When the Browser doesnot Support this video type-->
         </video>
         ```  

         * **< source > :** acts as an alternative media source provider for elements like < video >
           
             * **src :** specifies the URL or path to the media file

             * **type :** defines the MIME type of the media file ("video/mpeg" for MP4) 

     * **Main Attributes :**

         * **width & hieght :** specify the dimensions of the video player

         * **controls :**  adds a default playback interface with controls like play, pause, volume, etc

         * **autoplay :**  allows the video to start playing automatically when the web page loads

         * **loop :** makes the video play continuously, restarting after it finishes

         * **muted :** mutes the video by default        

         * **poster :** Specifies an image to be displayed before the video starts or has been paused 

             ```html
             <video poster="Image Path">
             ```    

     * **< track > :**      

         * Used to add timed text tracks to < video > and < audio > elements. These timed text tracks can include subtitles, captions, descriptions     

         * **Main Attributes :**

             * **kind :** 

                 * specifies the type of text track , It Can Be :

                     1) **subtitles :** Text that represents dialogue or narration

                     2) **captions :** Text that describes sounds, sound effects, and other non-speech audio content. (For the hearing impaired)

                     3) **descriptions :** Text that describes the visual content of the video for visually impaired users

             * **src :** 

                 * specifies the path to the text track file containing the timed text data. The file format is typically WebVTT **(.vtt files)**

# **Form :** 

 * Allows users to input data. This data can be text, choices from options, files, or anything else defined by the form

 * Collected data is then submitted to a web server for processing

 * **< form >** tag defines the start and end of the form

 * **Basic Structure :**

   ```html
   <form action="" method="">
    <input type="">
   </form>
   ```  

 * **< form > Atrributes :**

   * **action :** 
   
     * specifies the URL of the web page or script that will process the form data (often a server-side script like PHP)

   * **method :**

     * defines how the form data is sent to the server

       * **GET :** Data is appended to the URL as a query string

       * **POST :** Data is sent as part of the HTTP request body    

   * **novalidate :**

     * specifies that browser-side validation should not be performed on the form elements when the form is submitted

   * **target :** 

     * specifies where the response to the form submission should be displayed

     * By default, if it is not specified in a form, the form's response will be loaded in the same window or tab where the form resides  


 * **< form > Elements :**

   * **< label > :**

     * Used to Connect a text label to a form element  

       ```html
       <label for="ID">Label Text<label> 
       ```

   * **< input > :** 

     * A Self-closing Tag Used to create various input controls where users can enter and submit data 

     * **Main Attributes :**

       * **type :**

         * Specifies the type of input control , It Could Be :

           * **Password :** 
           
             * Creates a password field where characters are masked for security  

           * **text :** 
           
             * Creates a single-line text field for textual input (default)

           * **submit :** 
           
             * Creates a submit button, which triggers the submission of form data to a server-side script for processing when clicked

           * **email :** 
           
             * Creates an input field for email addresses. It offers basic validation to check for a valid email format 

           * **hidden :** 
           
             * Creates a hidden input field within a form , It Stores additional data that needs to be submitted with the form but isn't relevant for users to see or interact with directly

           * **reset :** 
           
             * Creates a button that resets all the form elements to their default values when clicked

           * **color :** 
           
             * Creates an input field that launches a color picker allowing users to select a color. The value is submitted as a hexadecimal color code 

           * **range :** 

             * Creates a slider input for selecting a value from a specified range

             * Attributes **(min - max - step)** to define the range and stepping intervals

           * **number :** 

             * Creates an input field for numeric values

             * Attributes **(min - max - step)** to specify allowed values and stepping intervals

           * **radio :** 

             * Creates radio buttons where users can select only one option from a group

           * **checkbox :**  

             * Creates checkboxes where multiple options can be selected

           * **file :**  

             * Creates a file upload field

           * **search :** 

             * Creates an input field specifically designed for search queries

           * **url :**

             * Creates an input field for URLs. Similar to **email**           


       * **id :** 

         * Assigns a unique identifier to the input element, useful for labeling and scripting

       * **name :** 

         * assigns a name to the input element. This name is used to reference the submitted data by the server-side script processing the form

       * **value :**

         * Sets the default value displayed in the input field  

       * **required :**

         * specifies that the input field is required and the user must enter a value before submitting the form

       * **placeholder :**

         * Specifies a hint or placeholder text displayed within the input field that disappears when the user starts typing

       * **readonly :**

         * allows users to see the value but prevents them from changing it

       * **disabled :**

         * disables the input field, making it uneditable for the user

       * **autofocus :** 

         * specifies that the particular element should automatically receive focus when the user loads the web page

       * **minlength & maxlength :**

         * Specify the allowed length (number of characters) for the user's input in a text field  

       * **checked :**

         * used with the **< input >** tag of type **checkbox** and **radio** to specify the initial checked state of the button    

       

   * **< select > :**  

     * Creates a dropdown list element that allows users to choose from a predefined set of options


     * **Basic Structure :**

       ```html
       <select name="" id="" multiple>
        <option value=""></option>               
        <option value=""></option>
       </select>
       ```

     * **< select attributes > :**

       * **name :**  Assigns a name to the dropdown list, used for referencing the selected value during form submission

       * **id :**  Uniquely identifies the dropdown list element for styling or scripting purposes

       * **multiple :** Allow users to select multiple options from the list   

     * **< option > :** 

       * define individual options that users can choose from in a dropdown list

       * **< option > attributes :**

         * **value :**  

           * specifies the value that is submitted along with the form when the user selects this option  

           * If not specified, the text content of the option is submitted by default

         * **selected :** 

           * used to pre-select a particular option as the default choice when the dropdown list loads    

     * **< opgroup > :**

       * Creates a group of options that are visually separated and labeled within a dropdown list   

       * **Example :**

         ```html
         <select name="" id="">
           <optgroup label="Numbers">
             <option value="">1</option>
             <option value="">2</option>
           </optgroup>
           <optgroup label="Characters">
             <option value="">A</option>
             <option value="">B</option>
           </optgroup>
         </select>
         ```

           * **label :** Specifies the text label displayed for the option group, helping users identify the category


   * **< textarea > :** 

     * creates a multi-line text input field. This is useful for situations where users need to enter extended text passages, comments, or descriptions within a web form   

     * **< textarea > Attributes :**

       * **name :** assigns a name to the textarea element, used for referencing the submitted text data on the server-side script

       * **id :**  Assigns a unique identifier to the textarea element for styling or scripting purposes

       * **cols :** specifies the number of columns displayed for the textarea, affecting its width

       * **rows :** specifies the number of visible lines displayed for the textarea in the web page

       * **placeholder :** Specifies a hint or placeholder text displayed within the textarea field that disappears when the user starts typing   


   * **< datalist > :**

     * works with the **< input list="">** element to provide an auto-completion suggestion list as the user types     

     * **Example :**

       ```html
       <input list="Languages" name="langs" placeholder="Select Your Language"/>
       <datalist id="Languages">
         <option>Arabic</option>
         <option>English</option>
         <option>French</option>
         <option>Dutch</option>
         <option>Italian</option>
       </datalist>
       ```
